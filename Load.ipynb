```python

from Transform import Transform
from url_banco_de_dados import url_criar_db, url_db
import psycopg2
import pandas as pd
import logging
from sqlalchemy import create_engine

def criar_banco_de_dados(url_cria_tabela):
    try:
        conn = psycopg2.connect(url_cria_tabela)
        conn.autocommit = True
        cursor = conn.cursor()

        cursor.execute(f"SELECT 1 FROM pg_catalog.pg_database WHERE datname = 'tabela_filmes';")
        if cursor.fetchone():
            logging.info("O banco de dados 'tabela_filmes' já existe.")
        else:
            cursor.execute("CREATE DATABASE tabela_filmes;")
            logging.info("Banco de dados 'tabela_filmes' criado com sucesso.")

        cursor.close()
        conn.close()
    except Exception as e:
        logging.error(f"Erro ao conectar ao PostgreSQL: {e}")

def Load(url_banco_de_dados, url_criar):

    criar_banco_de_dados(url_criar)
    
    engine = create_engine(url_banco_de_dados)

    try:
        df = Transform()  
        
        df.to_sql('tabela_filmes', con=engine, if_exists='replace', index=False)
        logging.info("Tabela criada ou substituída com sucesso.")
    except Exception as e:
        logging.error(f"Erro ao criar ou substituir a tabela: {e}")
    
    return pd.read_sql("SELECT * FROM tabela_filmes", engine)

print(Load(url_db, url_criar_db))
```
